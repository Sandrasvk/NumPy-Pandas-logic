# Group by

```python
import pandas as pd
```

```python
data = {'company' : ['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
        'Persons' : ['SAM', 'CHARLIE', 'AMY', 'CARL', 'JOE', 'SARAH'],
         'Sales' : [200, 120, 340, 124, 243, 350]}
                     
```

```python
df = pd.DataFrame(data)
df
```

```python
str = df.groupby('company')
```

```python
str.mean(numeric_only = True)
```

```python
df.groupby('company').mean(numeric_only = True)
```

```python
df.groupby('company').std(numeric_only = True)
```

```python
str.std(numeric_only = True)
```

```python
str.max()
```
