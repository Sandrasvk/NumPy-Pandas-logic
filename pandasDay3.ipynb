```python
#merging, joining and concatenation
df1 = pd.DataFrame({'A' : [ 'A0', 'A1', 'A2', 'A3'],
                    'B' : ['B0',  'B1', 'B2', 'B3'],
                    'C' : ['C0', 'C1', 'C2', 'C3']},
                     index = [0, 1, 2, 3])
 
```

```python
df2 = pd.DataFrame({'A' : ['A4', 'A5' ,'A6', 'A7'],
                   'B' : ['B4', 'B5', 'B6', 'B7'],
                   'C' : ['C4', 'C5', 'C6', 'C7']},
                    index = [4, 5, 6, 7])
```

```python
df3 = pd.DataFrame({'A' : ['A8', 'A9' ,'A10', 'A11'],
                   'B' : ['B8', 'B9', 'B10', 'B11'],
                   'C' : ['C8', 'C9', 'C10', 'C11']},
                    index = [8, 9, 10, 11])
```

```python
#merging/joining
left = pd.DataFrame({'key1' : ['k0', 'k0', 'k1', 'k2'],
                     'key2' : ['k0', 'k1', 'k0', 'k1'],
                       'A'  :  ['A0', 'A1', 'A2', 'A3'],
                       'B'  :  ['B0', 'B1', 'B2', 'B3']})
                               
```

```python
right = pd.DataFrame({'key1' : ['k0', 'k1', 'k1', 'k2'],
                     'key2' : ['k0', 'k0', 'k0', 'k0'],
                       'C'  :  ['C0', 'C1', 'C2', 'C3'],
                       'D'  :  ['D0', 'D1', 'D2', 'D3']})
                               
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, how = 'outer', on = ['key1', 'key2'])
```

```python
pd.merge(left, right, how = 'right', on = ['key1', 'key2'])
```
