Write a program to stack given 2 series horizonatally and vertically

```python
s1 = pd.Series([1, 2, 3])
s2 = pd.Series([4, 5, 6])
h_stack = pd.concat([s1, s2], axis = 1)
h_stack
```

Write a program to display all sunday of 2020:

```python
import pandas as pd
sundays = pd.date_range(start = '2020-01-01', end = '2020-12-31', freq = 'W-SUN')

for day in sundays:
    print(day.date())
```
